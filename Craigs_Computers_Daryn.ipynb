```python
#Import libraries
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#load craigs list
browser = Browser('chrome')
url = "https://kansascity.craigslist.org/search/sya"
browser.visit(url)
```

```python
# Create the Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract the elements
elems = soup.find_all('div', class_='gallery-card')
```

```python
# Get the data and put it in a list
craigs_data = []
for elem in elems:
    try: 
        price = elem.find('span', class_='priceinfo').text
    except:
        price = '$0'
    item_title = elem.find('span', class_='label').text
    meta = elem.find('div', class_='meta').text
    meta_list = meta.split("·")
    time_listed = meta_list[0]
    location = meta_list[1]
    craigs_data.append({'List_Type': 'Computers','Price': price, 'Item_Title': item_title, 'Time_Listed': time_listed, 'Location': location})
    
```

```python
df = pd.DataFrame.from_dict(craigs_data)
df
```

```python
df.to_csv('Output/craigs_computer_data.csv', index=False)
```
