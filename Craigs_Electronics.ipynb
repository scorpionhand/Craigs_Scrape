```python
import pandas as pd
from bs4 import BeautifulSoup
from splinter import Browser
```

```python
browser = Browser('chrome')
```

```python
browser.visit("https://kansascity.craigslist.org/search/ela#search=1~gallery~0~0")
```

```python
electronics_soup = BeautifulSoup(browser.html,'html.parser')
```

```python
electronics = electronics_soup.find_all('div', class_='gallery-card')
```

```python
price = electronics_soup.find_all(class_="priceinfo")
```

```python
electronics_data = []
for electronic in electronics:
    try:
        price = electronic.find('span', class_='priceinfo').text
    except:
        price = "$0"
    item_title = electronic.find('span', class_='label').text
    meta = electronic.find('div', class_='meta').text
    meta_list = meta.split("Â·")
    time_listed = meta_list[0]
    location = meta_list[1]
    electronics_data.append({'List_Type': 'Computers','Price': price, 'Item_Title': item_title, 'Time_Listed': time_listed, 'Location': location})
    
```

```python
electronics_df = pd.DataFrame.from_dict(electronics_data)
electronics_df.head(20)
```

```python
electronics_df.to_csv('Output/craiglist_electronics.csv', index=False)
```
